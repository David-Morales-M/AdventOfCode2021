# Parte 1

```python
# Abrimos el archivo de datos
with open("Data.txt") as f:

  # Variable del Cardumen
  shoal = ()
  for line in f:
    shoal = line.split(',')

  # En el rango de 80 días
  for j in range(80):
    # Analizamos cada pez
    for i in range(len(shoal)):
      # Si este tiene 0 días, su espera para reproducirse será 6 días
      # Se agrega otro pez con espera para reproducirse de 8 días
      if (int(shoal[i])==0):
        shoal[i] = '6'
        shoal.append(8)
      # Si no pasa nada, se procede a disminuir un día al tiempo de reproducción
      else:
        shoal[i] = str(int(shoal[i]) - 1)

# Resultados
print(shoal)
print('Longitud del Cardumen:' + str(len(shoal)))
```

# Parte 2

```python
# Realizar con el método anterior es computacionalmente enorme
# procedemos a cambiar de pensamiento

# Abrimos el archivo de datos
with open("Data.txt") as f:

  # Variable del Cardumen
  shoal = ()
  fishes = {}
  sum = 0
  accum = 0
  for line in f:
    shoal = line.split(',')

  # Transformando la lista en un diccionario con la cantidad 
  # de peces según su día
  for i in range(9):
    fishes[(i)] = shoal.count(str(i))

  # Dados los 256 días
  for j in range(257):
    # Si los hay peces en el día 0
    # Estos se suman a los peces del día 8 y 6
    # Los peces del día i, serán los peces del día i+1
    if fishes[0]>0:
      accum = fishes[0]
      fishes[0] = fishes[1]
      fishes[1] = fishes[2]
      fishes[2] = fishes[3]
      fishes[3] = fishes[4]
      fishes[4] = fishes[5]
      fishes[5] = fishes[6] 
      fishes[6] = fishes[7] + accum
      fishes[7] = fishes[8]
      fishes[8] = accum
    else: 
      accum = fishes[0]
      fishes[0] = fishes[1]
      fishes[1] = fishes[2]
      fishes[2] = fishes[3]
      fishes[3] = fishes[4]
      fishes[4] = fishes[5]
      fishes[5] = fishes[6] 
      fishes[6] = fishes[7] 
      fishes[7] = fishes[8]
      fishes[8] = accum

  print(fishes)

  # Contamos la cantidad de peces finales
  for i in range(8):
    sum = sum + fishes[i]

# Resultados
print('Longitud del Cardumen:' + str(sum))
```
