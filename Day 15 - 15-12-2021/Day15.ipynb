# Parte 1

```python
# La solución consiste en crear un grafo con los datos de entrada
# para emplear el algoritmo de Dijkstra y encontrar el camino más
# corto

# Ingresando todos los datos en una matriz de enteros
data = np.genfromtxt('Data.txt', delimiter=1, dtype=int)
# Creamos un grafo vacio    
grafo = nx.DiGraph()

# Implementamos el iterador N-dimensional, según la dimensión de la matriz
for x, y in np.ndindex(data.shape):
    # Si x es mayor a 0, añadimos un camino entre un nodo (x-1, y) y el otro (x,y)
    # especificando su peso como el valor de la posición (x,y)
    if x > 0: 
        grafo.add_edge((x - 1, y), (x, y), weight = data[(x, y)])

    # Si x es menor a la dimensión horizontal (x-10) de la matriz, añadimos un 
    # camino entre un nodo (x-1, y) y el otro (x,y) especificando su peso como 
    # el valor de la posición (x,y)
    if x < data.shape[0] - 1:
        grafo.add_edge((x + 1, y), (x, y), weight = data[(x, y)])

    # Lo mismo realizado con x se aplica en y.
    if y > 0:
        grafo.add_edge((x, y - 1), (x, y), weight = data[(x, y)])

    if y < data.shape[1] -1:
        grafo.add_edge((x, y + 1), (x, y), weight = data[(x, y)])

# Resultado
# Una vez construido el grafo, empleamos la función "dijkstra_path_length" de la
# librería networkx, con la cual determinaremos la ruta más corta desde el origen 
# (0,0) al final del grafo (dimensión[x] -1) (dimension[y] - 1). 
print('Coste del camino más corto: ' + str(nx.dijkstra_path_length(grafo, (0, 0), (data.shape[0] - 1, data.shape[1] -1))))
```

```python
# Cuando el mapa es 5 veces más grande debemos generar una matriz más grande


# Ingresando todos los datos en una matriz de enteros
data = np.genfromtxt('Data.txt', delimiter=1, dtype=int) 
# Ahora procederemos a unir los arreglos de cada eje 5 veces  
# En x
data = np.concatenate((data, data + 1, data + 2, data + 3, data + 4), axis=0)
# En y
data = np.concatenate((data, data + 1, data + 2, data + 3, data + 4), axis=1)

# Si el dato en la matriz es mayor a 9, le restamos 9, para reiniciarlo al valor
# 1
data[data > 9] -= 9
# Creamos un grafo vacio    
grafo = nx.DiGraph()

# Implementamos el iterador N-dimensional, según la dimensión de la matriz
for x, y in np.ndindex(data.shape):
    # Si x es mayor a 0, añadimos un camino entre un nodo (x-1, y) y el otro (x,y)
    # especificando su peso como el valor de la posición (x,y)
    if x > 0: 
        grafo.add_edge((x - 1, y), (x, y), weight = data[(x, y)])

    # Si x es menor a la dimensión horizontal (x-10) de la matriz, añadimos un 
    # camino entre un nodo (x-1, y) y el otro (x,y) especificando su peso como 
    # el valor de la posición (x,y)
    if x < data.shape[0] - 1:
        grafo.add_edge((x + 1, y), (x, y), weight = data[(x, y)])

    # Lo mismo realizado con x se aplica en y.
    if y > 0:
        grafo.add_edge((x, y - 1), (x, y), weight = data[(x, y)])

    if y < data.shape[1] -1:
        grafo.add_edge((x, y + 1), (x, y), weight = data[(x, y)])

# Resultado
# Una vez construido el grafo, empleamos la función "dijkstra_path_length" de la
# librería networkx, con la cual determinaremos la ruta más corta desde el origen 
# (0,0) al final del grafo (dimensión[x] -1) (dimension[y] - 1). 
print('Coste del camino total más corto: ' + str(nx.dijkstra_path_length(grafo, (0, 0), (data.shape[0] - 1, data.shape[1] -1))))
```
